```python
from rossby_waves import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Amplitude and Dispersion

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
X, Y = np.meshgrid(x, y)
amp = amplitude([X, Y])
dis = dispersion([X, Y])
```

```python
fig = plt.figure(figsize=(10, 4))
fig.suptitle("Amplitude")

ax1 = fig.add_subplot(1, 2, 1)
plot = ax1.contourf(X, Y, amp, 60, cmap="coolwarm")
ax1.set_xlabel("k")
ax1.set_ylabel("l")
cbar = fig.colorbar(plot, pad=0.05)
cbar.ax.set_ylabel("amplitude, m")

ax2 = fig.add_subplot(1, 2, 2, projection="3d")
ax2.contourf3D(X, Y, amp, 100, cmap="coolwarm")
ax2.set_xlabel("k")
ax2.set_ylabel("l")
ax2.set_zlabel("amplitude, m");
```

```python
# plt.contourf(X, Y, dis, 50)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, dis, 500)
ax.view_init(20, 45)
ax.set_xlabel("k")
ax.set_ylabel("l")
ax.set_zlabel("omega, 1/t")
ax.set_title("Dispersion Relation");
```
