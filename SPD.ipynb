```python
from rossby_waves import *
from math import floor
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from rossby_waves.rossby_waves import amplitude, dispersion, grid, trajectory, vel, vel_autocor, trajectory_den, traj2, dt2
from matplotlib.animation import FuncAnimation, PillowWriter
```

```python
r = RossbyWave(np.array([5, 5]))
ro = RossbyOcean([r])
ro.add_grid_waves()
ro.remove_wave(0)
x = grid(100)
t = np.linspace(0, 5e12,2001)
```

Divergence Free Flow (eps = 0)

```python
#in x direction
xcoords, ycoords = traj2(ro,x,0,5e12, 2.5e9, 0)
p1= dt2(xcoords)
fig = plt.figure(figsize=(45, 10))
ax = fig.add_subplot(111)
ax.plot(t,p1)
ax.grid(True, which='both')
ax.set_ylim([0,3])
ax.set_ylabel("$D_x(t)$",fontsize=40)
ax.set_xlabel("Time,$(10^{12})$",fontsize=40)
ax.tick_params(axis='x', labelsize=40)
ax.tick_params(axis='y', labelsize=40)
fig.show
plt.savefig("spd0x.png")
```

```python
#in y direction
p1= dt2(ycoords)
fig = plt.figure(figsize=(45, 10))
ax = fig.add_subplot(111)
ax.plot(t,p1)
ax.grid(True, which='both')
ax.set_ylim([0,3])
ax.set_ylabel("$D_y(t)$",fontsize=40)
ax.set_xlabel("Time,$(10^{12})$",fontsize=40)
ax.tick_params(axis='x', labelsize=40)
ax.tick_params(axis='y', labelsize=40)
fig.show
plt.savefig("spd0y.png")
```
