```python
r = RossbyWave(np.array([1, 1]))
ro = RossbyOcean([r])
ro.add_random_waves(10, beta=10)
ro.remove_wave(0)
ro.plot_streamfunction()
```

```python
x0 = np.array([1,1])
```

```python
ro.animate_trajectory(x0, tlim=(0, 1, 100), solenoidal = True)
```

```python
x_, y_ = trajectories(r, [np.array([1,0]), np.array([0,1])], 0, 1, 10)
x_[1][0]
```

```python
points = grid(3)
ro.animate_trajectories(points)
```

```python
points = grid(50)
endpoints = []
for x0 in points:
    x, y = trajectory(ro, x0, 0, 5, 100, solenoidal = True)
    endpoints.append(np.array([x[-1], y[-1]]))
endpoints
```
